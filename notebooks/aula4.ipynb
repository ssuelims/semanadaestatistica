# Aula 2

```python
plt.figure(figsize=(9,3))
# 1 polegada = 2.54 cm
# largura 8 = 20.32 cm
# altura 4 = 10.16 cm
# DPI (dots per inch, ou pontos por polegada).
# padrao no matplotlib - DPI = 100
# Largura : 8 polegadas x 100 DPI = 800 pixels.
# Altura : 4 polegadas x 100 DPI = 400 pixels


plt.plot(dados_brasil['ano'],dados_brasil['imigrantes'])

plt.xticks(['1980','1985','1990','1995','2000','2005','2010'])

plt.yticks([500,1000,1500,2000,2500,3000])
plt.grid(linestyle='--')
plt.title('imigraçao do brasil para o canadá')
plt.xlabel('ano')
plt.ylabel('numero de imigrantes')
plt.show()
```

```python
# criando  uma figura
fig,axs = plt.subplots(1,2,figsize = (15,5))

# Subplot 1: Grafico de linha

axs[0].plot(dados_brasil['ano'],dados_brasil['imigrantes'])
axs[0].xaxis.set_major_locator(plt.MultipleLocator(5))
axs[0].set_xlabel('ano')
axs[0].set_ylabel('numero de imigrantes')
axs[0].set_title('imigracao do Brasil para o canada')

# Subplot 2: Grafico Boxplot
axs[1].boxplot(dados_brasil['imigrantes'])
axs[1].set_title('Boxplot da imigraçao do Brasil para o canada')
axs[1].set_xlabel('Brasil')
axs[1].set_ylabel('Numero de imigrantes')
axs[1].grid()

plt.show()
```

# Comparativo de imigracoes do brasil com alguns paises da America Latina

```python
# Criar uma figura 
fig,axs = plt.subplots(2,2,figsize=(10,6))
fig.subplots_adjust(hspace=0.5,wspace=0.3)

# Subplots
# Brasil

axs[0,0].plot(df.loc['Brasil',anos])
axs[0,0].set_title('brasil')
axs[0,0].set_xlabel("anos")
axs[0,0].set_ylabel("Quantidade de imigrantes")

# Colombia
axs[0,1].plot(df.loc['Colômbia',anos])
axs[0,1].set_title('colômbia')
axs[0,1].set_xlabel('anos')
axs[0,1].set_ylabel('quantidade de imigrates')

# Argentina
axs[1,0].plot(df.loc['Argentina',anos])
axs[1,0].set_title('Argentina')
axs[1,0].set_xlabel('anos')
axs[1,0].set_ylabel('quantidade de imigrantes')

# peru
axs[1,1].plot(df.loc['Peru',anos])
axs[1,1].set_title('Peru')
axs[1,1].set_xlabel('anos')
axs[1,1].set_ylabel('quantidade de imigrantes')


for ax in axs.flat:
    ax.xaxis.set_major_locator(plt.MultipleLocator(5))

  # Usando ravel() - achatando a matriz evitando o uso complexo do 
  # Usando set_ylim() - definir o limiteos valores
ymim = 0
ymax = 7000

for ax in axs.ravel():
    ax.set_ylim(ymim,ymax) 
   
```

# Aula 04 -30/08/2024

# deixando o grafico na horizontal

```python
# Criando a figura
fig, ax = plt.subplots(figsize=(12, 5))
# E um grafico de bar (Matplotlib)
ax.barh(america_sul.index, america_sul['Total'], color=cores)
# Titulo alinha a esquerda com fonte 18
ax.set_title('Imigração da América do Sul para o Canadá\n1980 a 2013', loc='left', fontsize=18)
# eixo y com fonte 14
ax.set_ylabel('', fontsize=14)
ax.set_xlabel('numero de imigrantes')
# Mudar as fontes dos ticks
ax.yaxis.set_tick_params(labelsize=12)
ax.xaxis.set_tick_params(labelsize=12)

# Insights
# storytelling
for i, (value, name) in enumerate(zip(america_sul['Total'], america_sul.index)):
    ax.text(value, i, f'{value:,}', va='center', ha='left', fontsize=12, color='black')

plt.show()
```
